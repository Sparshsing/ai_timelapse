```python
import cv2

# Read source images (must be same size)
img1 = cv2.imread('input_frames/boy/one.jpg')
img2 = cv2.imread('input_frames/boy/two.jpg')

# Choose blending factor α (0.0 → only img1, 1.0 → only img2)
alpha = 0.5  

# Compute weighted sum
blended = cv2.addWeighted(img1, 1-alpha, img2, alpha, 0)
```

```python
# Display the blended image using matplotlib
import matplotlib.pyplot as plt
plt.imshow(cv2.cvtColor(blended, cv2.COLOR_BGR2RGB))
plt.axis('off')  # Hide axes
plt.show()
```

```python
import os
os.makedirs('output_frames_cv', exist_ok=True)
# Save the blended image
cv2.imwrite('output_frames_cv/transition.jpg', blended)
```
